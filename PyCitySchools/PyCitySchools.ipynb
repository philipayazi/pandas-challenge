```python
import pandas as pd
```

```python
#Load data files
school_data_file = 'Resources/schools_complete.csv'
student_data_file = 'Resources/students_complete.csv'
```

District Summary Report

```python
#read data files and store into pandas data frames
school_df = pd.read_csv(school_data_file)

school_df
```

```python
school_df.info()
```

```python
student_df = pd.read_csv(student_data_file)

student_df.head()
```

```python
student_df.info()
```

```python
#merge data frames
complete_df = pd.merge(student_df, school_df, how='outer')
complete_df.head()
```

```python
#find total number of school, total number of students, total budget for district

total_schools = school_df.school_name.count()
total_students = complete_df.student_name.count()
total_dist_bdgt = school_df.budget.sum()

total_schools
total_students
total_dist_bdgt
```

```python
# find avg math score for district

sum_math_score = complete_df.math_score.sum()
avg_math_score = round(float(sum_math_score/total_students), 6)

avg_math_score
```

```python
# avg reading score for district, 


sum_reading_score = complete_df.reading_score.sum()
avg_reading_score = round(float(sum_reading_score/total_students), 6)

avg_reading_score
```

```python
#create new columns for students pass or fail for math and reading

complete_df['math_pass_or_fail'] = pd.cut(complete_df['math_score'],
                          bins=[0,70,101],
                          labels=['fail','pass'])
complete_df['reading_pass_or_fail'] = pd.cut(complete_df['reading_score'],
                                          bins=[0,70,101],
                                          labels=['fail','pass'])

complete_df.head()
```

```python
#find %passing math


passing_math_df = complete_df.loc[complete_df['math_pass_or_fail'] == 'pass']
total_passing_math = passing_math_df.math_pass_or_fail.count()
pct_passing_math = round(float((total_passing_math/total_students) * 100), 6)
pct_passing_math
```

```python
#find % passing reading

passing_reading_df = complete_df.loc[complete_df['reading_pass_or_fail'] == 'pass']
total_passing_reading = passing_reading_df.reading_pass_or_fail.count()
pct_passing_reading = round(float((total_passing_reading/total_students) * 100), 6)
pct_passing_reading
```

```python
#overall passing rate(average of the %passing math and reading)
pct_overall_passing = (pct_passing_math + pct_passing_reading)/2
pct_overall_passing
# overall_passing_df = passing_reading_df.loc[passing_reading_df['math_pass_or_fail'] == 'pass']
# overall_passing = overall_passing_df.math_pass_or_fail.count()
# pct_overall_passing = round(float((overall_passing/total_students) * 100), 2)
pct_overall_passing
```

```python
#Create District Summary
district_summary = pd.DataFrame([[total_schools, total_students, total_dist_bdgt, avg_math_score, avg_reading_score, pct_passing_math, pct_passing_reading, pct_overall_passing]],
                                index=[0],
                                columns=['Total Schools', 'Total Students', 'Total Budget', 'Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing Rate']
)

district_summary
```

```python
#count students per school
student_count = complete_df.groupby('school_name')['math_pass_or_fail'].count()
```

```python
school_df_with_grades = pd.merge(school_df_with_grades, pct_passedM.rename('Percent Passing math'), on='school_name')

school_df_with_grades
```

```python
# accidently ran the code twice on previous line and added an extra percent passing math column. Needed to fix this.
school_df_with_grades = school_df_with_grades.drop(['Percent Passing math_x', 'Percent Passing math_y'], axis=1)
school_df_with_grades = school_df_with_grades.rename(columns={'Percent Passing math_x':'Percent Passing Math'})
school_df_with_grades
```

```python
#filter students that have passed reading
passed_students_read = complete_df.loc[complete_df['reading_pass_or_fail'] == 'pass']
# and then group those students by school and count
passed_students_count_read = passed_students_read.groupby('school_name')['reading_pass_or_fail'].count()
```

```python
# find the percent of students passing reading
pct_passedR = round(passed_students_count_read/student_count * 100, 6)
pct_passedR
```

```python
# merge the series to the dataframe
school_df_with_grades = pd.merge(school_df_with_grades, pct_passedR.rename('Percent Passing Reading'), on='school_name')

school_df_with_grades
```

```python
# find overall average passing rate
overall_pass_rate = (pct_passedM + pct_passedR)/2
overall_pass_rate
```

```python
# merge this series to the dataframe
school_df_with_grades = pd.merge(school_df_with_grades, overall_pass_rate.rename('Percent Overall Passing Rate'), on='school_name')
school_df_with_grades
```
